### Density estimator in PRML 2.5. Dataset comes from siemens data challenge.

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
with open("trainingdata_a.csv", "r") as file:
    data = csv.reader((line.replace('\0','') for line in file), delimiter=",")
    dataset = [line[1:] for index, line in enumerate(data) if index != 0]
dataset = np.array(dataset, dtype="float")
X_data, Label = dataset[:,:2], dataset[:,-1][:, np.newaxis]

index = [i for i, item in enumerate(Label==0) if item==True]
XClass0 = X_data[index]
YClass0 = Label[index]
print("Class 0 has shape: ", XClass0.shape)
index = [i for i, item in enumerate(Label==1) if item==True]
XClass1 = X_data[index]
YClass1 = Label[index]
print("Class 1 has shape: ", XClass1.shape)

plt.scatter(XClass0[:,0], XClass0[:,1], c='b')
plt.scatter(XClass1[:,0], XClass1[:,1], c='g')
```

### 1. Kernel density estimators, P122. I calculated the probability first then normalize with sum.

```python
# 2.248
def panzer_prob(x, dataset, h):
    K = 0
    N = len(dataset)
    D = dataset.shape[1]
    V = h**D
    temp = (x - dataset) / h
    for i in range(temp.shape[0]):
        if np.abs(temp[i][0])<=0.5 and np.abs(temp[i][1])<=0.5:
            K += 1
    prob = K/(N*V)
    return prob
    
# test, the probability of x appear in (0.5, 0.5) with h=0.4
print(panzer_prob(np.array([0.5,0.5]), XClass0, 0.4))

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, subplot_kw={"projection": "3d"})
f.set_figheight(15)
f.set_figwidth(15)
X = np.arange(0, 1, 0.05)
Y = np.arange(0, 1, 0.05)
X, Y = np.meshgrid(X, Y)
Z1 = np.zeros((X.shape))

sum1 = 0.
for i in range(Z1.shape[0]):
    for j in range(Z1.shape[1]):
        x = np.array([X[i,j],Y[i,j]]).reshape(1,2)
        Z1[i,j] = panzer_prob(x, XClass0, h=0.4)
        sum1 += Z1[i,j]
Z1 /= sum1
        
sum2 = 0.     
Z2 = np.zeros((X.shape))
for i in range(Z2.shape[0]):
    for j in range(Z2.shape[1]):
        x = np.array([X[i,j],Y[i,j]]).reshape(1,2)
        Z2[i,j] = panzer_prob(x, XClass1, h=0.4)
        sum2 += Z2[i,j]
Z2 /= sum2

surf1 = ax1.plot_surface(X, Y, Z1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z2, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

### 2. Gaussian Kernel, P123. 

```python
def Gaussian_prob(x, dataset, h):
    K = 0
    N = len(dataset)
    D = dataset.shape[1]
    V = (2*np.pi*h**2)**D/2
    
    for data in dataset:
        inner_exp = - np.linalg.norm(x - data)**2/(2*h**2)
        K += np.exp(inner_exp)
    prob = K/(N*V)
    return prob

# test
print(Gaussian_prob(np.array([0.5,0.5]), XClass0, 0.3))

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, subplot_kw={"projection": "3d"})
f.set_figheight(15)
f.set_figwidth(15)
X = np.arange(0, 1, 0.05)
Y = np.arange(0, 1, 0.05)
X, Y = np.meshgrid(X, Y)
Z1 = np.zeros((X.shape))

sum1 = 0.
for i in range(Z1.shape[0]):
    for j in range(Z1.shape[1]):
        x = np.array([X[i,j],Y[i,j]]).reshape(1,2)
        Z1[i,j] = Gaussian_prob(x, XClass0, h=0.1)
        sum1 += Z1[i,j]
Z1 /= sum1
        
sum2 = 0.
Z2 = np.zeros((X.shape))
for i in range(Z2.shape[0]):
    for j in range(Z2.shape[1]):
        x = np.array([X[i,j],Y[i,j]]).reshape(1,2)
        Z2[i,j] = Gaussian_prob(x, XClass1, h=0.1)
        sum2 += Z2[i,j]
Z2 /= sum2

surf1 = ax1.plot_surface(X, Y, Z1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z2, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

### We could see, gaussian kernel is smoother than panzer kernel. You could also try to tune h parameter to get more intuition. 

### 3. Neareat neighbour methods P124, super discontinuous boundary

```python
# implement 2.256 p(Ck|x)

def KNN_prob(x, dataset, k, Ck):
    temp = np.linalg.norm(x - dataset, axis=1)
    index = np.argsort(temp)
    K_index = index[:k]
    classes = Label[K_index]
    Kk = np.sum(classes == Ck)
    return Kk/len(dataset)

# test
KNN_prob(np.array([0.5,0.5]), X_data, 100, Ck=0)

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, subplot_kw={"projection": "3d"})
f.set_figheight(15)
f.set_figwidth(15)
X = np.arange(0, 1, 0.05)
Y = np.arange(0, 1, 0.05)
X, Y = np.meshgrid(X, Y)
Z1 = np.zeros((X.shape))

sum1 = 0.
for i in range(Z1.shape[0]):
    for j in range(Z1.shape[1]):
        x = np.array([X[i,j],Y[i,j]]).reshape(1,2)
        Z1[i,j] = KNN_prob(x, X_data, k=50, Ck=0) # class 0, with k = 50
        sum1 += Z1[i,j]
Z1 /= sum1
        
sum2 = 0.
Z2 = np.zeros((X.shape))
for i in range(Z2.shape[0]):
    for j in range(Z2.shape[1]):
        x = np.array([X[i,j],Y[i,j]]).reshape(1,2)
        Z2[i,j] = KNN_prob(x, X_data, k=50, Ck=1) # class 1, with k = 50
        sum2 += Z2[i,j]
Z2 /= sum2

surf1 = ax1.plot_surface(X, Y, Z1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
surf2 = ax2.plot_surface(X, Y, Z2, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```
